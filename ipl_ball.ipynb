```python
# Import necessary libraries
import numpy as np
import matplotlib.pyplot as plt

# Load deliveries dataset
file_path = "../datasets/deliveries.csv"  # Adjust path if needed
data = np.genfromtxt(file_path, delimiter=',', dtype=str, skip_header=1)

# Display basic info
print(f"Dataset Loaded Successfully with shape: {data.shape}")
```

```python
# Analyze Most Common Dismissal Types
dismissal_kinds = data[:, 17]  # Column 18 → dismissal_kind
unique_dismissals, dismissal_counts = np.unique(dismissal_kinds[dismissal_kinds != ''], return_counts=True)

print("Most Common Dismissal Types:")
for dismissal, count in zip(unique_dismissals, dismissal_counts):
    print(f"{dismissal}: {count}")
```

```python
# Calculate Total Extras Runs
extras_runs = data[:, 16].astype(int)  # Column 17 → extra_runs
total_extras = np.sum(extras_runs)

print(f"\nTotal Extra Runs Given in the Tournament: {total_extras}")
```

```python
# Plot Total Runs Per Over
overs = data[:, 7]  # Column 8 → over number
total_runs = data[:, 15].astype(int)  # Column 16 → total_runs

# Group runs by over
over_nums = np.unique(overs)
runs_per_over = [np.sum(total_runs[overs == over]) for over in over_nums]

plt.figure(figsize=(12, 6))
plt.bar(over_nums, runs_per_over, color='skyblue')
plt.xlabel("Over")
plt.ylabel("Total Runs Scored")
plt.title("Total Runs Scored Per Over")
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```
