```python
import numpy as np
import matplotlib.pyplot as plt
import csv

```

```python
file_path = "../datasets/matches.csv"

teams = []

with open(file_path, "r", encoding='utf-8') as file:
    reader = csv.reader(file)
    header = next(reader)  # Skip header row
    winner_index = header.index("winner")  # Find the index of 'winner' column
    
    for row in reader:
        teams.append(row[winner_index])  # Store match winners

print("Dataset Loaded Successfully ✅")
```

```python
teams = np.array(teams)  # Convert to numpy array
teams = teams[teams != '']  # Remove empty values (abandoned matches)

print(f"Total Matches With Winners: {len(teams)}")
```

```python
unique_teams, wins = np.unique(teams, return_counts=True)

# Display Team Wins
print("\nWins Per Team:")
for team, win in zip(unique_teams, wins):
    print(f"{team}: {win} wins")
```

```python
sorted_indices = np.argsort(wins)[::-1]  # Sort by wins (descending order)
unique_teams = unique_teams[sorted_indices]
wins = wins[sorted_indices]

# Display Sorted Wins
print("\nTeams Sorted By Wins:")
for team, win in zip(unique_teams, wins):
    print(f"{team}: {win} wins")
```

```python
plt.figure(figsize=(12, 6))
plt.bar(unique_teams, wins, color='skyblue')
plt.xticks(rotation=45, ha='right')
plt.title("IPL Wins Per Team")
plt.ylabel("Number of Wins")
plt.tight_layout()
plt.show()
```
